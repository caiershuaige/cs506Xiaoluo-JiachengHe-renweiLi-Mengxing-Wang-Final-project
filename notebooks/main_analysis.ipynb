```python
# 1. imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 2. load data
df = pd.read_csv("../violations.csv")

# 3. clean data
df = df.dropna(subset=["description", "status_dttm", "violation_city"])
df["date"] = pd.to_datetime(df["status_dttm"], errors="coerce")
df = df.dropna(subset=["date"])
df["year"] = df["date"].dt.year
df["violation_type"] = df["description"].str.lower().str.strip()
df["neighborhood"] = df["violation_city"].str.strip()
```
